```python
import pandas as pd
```

```python
df = pd.read_csv('/nfs/data/COM2POSE/reference_data/gencode.vM25.annotation.gtf', sep='\t', header=None, comment='#')
df = df[df[2] == 'gene']
```

```python
# Split the 9th column by semicolon into distinct columns
df = df[8].str.split(';', expand=True)[[0, 2]]
df.columns = ['gene_id', 'gene_name']
```

```python
df["gene_id"] = df["gene_id"].str.lstrip('gene_id ').str.strip('"')
df["gene_name"] = df["gene_name"].str.lstrip('gene_name ').str.strip('"')
```

```python
df.index = df["gene_name"]
df.drop("gene_name", axis=1, inplace=True)
df["gene_id"] = df["gene_id"].str.split('.', expand=True)[0]

# Group by index, select first gene id
df = df.groupby(level=0).first()
```
